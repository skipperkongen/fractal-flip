# Fractals

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## One point set in 2D

```python
def oneset2d(ps, wn=2, n=20000):
    cur = ps[0]
    for _ in range(n):
        p = ps[np.random.choice(ps.shape[0], 1, replace=False), :][0]
        cur = (1/wn)*cur + (1-(1/wn))*p
        yield cur

def fractal(ps, wn=2, n=20000):
    trace = oneset2d(ps, n=n, wn=wn)
    trace = np.array([[x,y] for x,y in trace])
    sns.scatterplot(x=trace[:,0], y=trace[:,1], s=1)
    sns.scatterplot(x=ps[:,0], y=ps[:,1], color='red')
    #plt.title(f'{len(ps)} points, 1/{wn} weight')
    plt.show()
```

### 3 points

```python
ps = np.array([[0,0], [1,0], [0.5, 0.5],])
for wn in range(2,5):
    fractal(ps, wn=wn)
```

### 4 points

```python
ps = np.array([[0,0], [1,0], [1,1], [0,1]])

for wn in range(2,5):
    fractal(ps, wn=wn)
```

### 5 points

```python
ps = np.array([[0,0], [1,0], [1,1], [0,1], [1/2, 1/2]])
for wn in range(2,5):
    fractal(ps, wn=wn)
```

### 6 points

```python
ps = np.array([[0, 0.8], [1, 0.8], [0.5,3], [0, 2.2], [1, 2.2], [0.5, 0]])
for wn in range(2,6):
    fractal(ps, wn=wn)
```

### 17 random points

```python
ps = np.random.random((17,2))
for wn in range(2,6):
    fractal(ps, wn=wn)
```

```python
ps = np.random.random((7,2))
for wn in range(2,6):
    fractal(ps, wn=wn)
```
